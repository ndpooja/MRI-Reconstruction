```python
# Importing all the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('070623_2DSE_phantom_256x256.dat', sep="\s+", header=None)
data_real = np.array(data.iloc[:, 3])
kspace_real = np.reshape(data_real, (256, 256))
data_imag = np.array(data.iloc[:, 4])
kspace_imag = np.reshape(data_imag, (256, 256))

kspace_data = np.zeros((256, 256),dtype = np.complex_)
kspace_data.real = kspace_real
kspace_data.imag = kspace_imag
```

```python
plt.figure(figsize=(10,10))
plt.imshow(np.abs(kspace_data), cmap='gray')
plt.title("Kspace Data (Cartesian)")
```

```python
norm = "ortho"

def fft(x):
    return np.fft.fft2(x)

def ifft(x):
    return np.fft.ifft2(x)

phantom_img = np.fft.fftshift(ifft(np.fft.fftshift(kspace_data)))

plt.figure(figsize=(10,10))
plt.imshow(np.abs(phantom_img), cmap = 'gray')
plt.title("Cartesian Image")
```
